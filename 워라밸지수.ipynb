```python
home_zeros_scaled["워라밸지수"]=3
home_zeros_scaled
```

```python
work=home_scaled.copy()
work_1=work['평일직장근처체류시간']
work_2=work['휴일직장근처체류시간']
level_1q_1=work_1.quantile(0.25)
level_3q_1=work_1.quantile(0.75)
level_1q_2=work_2.quantile(0.25)
level_3q_2=work_2.quantile(0.75)
rev_range=3
IQR1=level_3q_1-level_1q_1
IQR2=level_3q_2-level_1q_2
condition1=(work_1<=level_3q_1+(rev_range*IQR1))&(work_1>=level_3q_1-(rev_range*IQR1))
condition2=(work_2<=level_3q_2+(rev_range*IQR2))&(work_2>=level_3q_2-(rev_range*IQR2))
total_work=work[condition1&condition2]
total_work
total_else=work[(condition1==False)|(condition2==False)]
```

```python
work_fit=total_work[['평일직장근처체류시간','휴일직장근처체류시간']].reset_index(drop=True)
work_fit
work_info=total_work[home_scaled.columns.difference(['평일직장근처체류시간','휴일직장근처체류시간'])].reset_index(drop=True)
work_info
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(work_fit)
work_scaled = scaler.transform(work_fit)
work_scaled_df = pd.DataFrame(data=work_scaled, columns=['평일직장근처체류시간','휴일직장근처체류시간'])
work_scaled_df
```

```python
wlv_making=pd.concat([work_info,work_scaled_df],axis=1)
wlv_making
```

```python
wlv_1=wlv_making.copy()
wlv_1['휴일가중치비율'][(wlv_making['휴일가중치비율']<=0.2)]=1.5
wlv_1['휴일가중치비율'][(wlv_making['휴일가중치비율']>0.2)]=2
wlv_1
```

```python
wlv_1['워라밸지수']=(1-wlv_1['평일직장근처체류시간'])*(1-wlv_1['휴일직장근처체류시간']*wlv_1['휴일가중치비율'])
wlv_1
```

```python
idx_wlv1=wlv_1['워라밸지수']
level_1q=idx_wlv1.quantile(0.25)
level_3q=idx_wlv1.quantile(0.75)
rev_range=3
IQR=level_3q-level_1q
condition=(idx_wlv1<=level_3q+(rev_range*IQR))&(idx_wlv1>=level_3q-(rev_range*IQR))
#condition=(idx_wlv<=20)&(idx_wlv>=-20)
idx_wlv1_df=wlv_1[condition][['기준시점','아파트단지코드','라이프스테이지','워라밸지수']]
idx_wlv1_df
```

```python
print(idx_wlv1_df['워라밸지수'].quantile(0.1))
print(idx_wlv1_df['워라밸지수'].quantile(0.3))
print(idx_wlv1_df['워라밸지수'].quantile(0.75))
print(idx_wlv1_df['워라밸지수'].quantile(0.9))
```

```python
wlv1=wlv_1.copy()
wlv1['워라밸지수'][(wlv_1['워라밸지수']<idx_wlv1_df['워라밸지수'].quantile(0.1))]=1
wlv1['워라밸지수'][(wlv_1['워라밸지수']>=idx_wlv1_df['워라밸지수'].quantile(0.1))&(wlv_1['워라밸지수']<idx_wlv1_df['워라밸지수'].quantile(0.3))]=2
wlv1['워라밸지수'][(wlv_1['워라밸지수']>=idx_wlv1_df['워라밸지수'].quantile(0.3))&(wlv_1['워라밸지수']<idx_wlv1_df['워라밸지수'].quantile(0.75))]=3
wlv1['워라밸지수'][(wlv_1['워라밸지수']>=idx_wlv1_df['워라밸지수'].quantile(0.75))&(wlv_1['워라밸지수']<idx_wlv1_df['워라밸지수'].quantile(0.9))]=4
wlv1['워라밸지수'][(wlv_1['워라밸지수']>=idx_wlv1_df['워라밸지수'].quantile(0.9))]=5
wlv1
```
