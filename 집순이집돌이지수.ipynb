####**1) 집순이집돌이지수**

```python
#집순이집돌이지수
```

```python
condition__1=(data.평일이동횟수!=0) & (data.평일이동거리!=0) & (data.휴일이동거리!=0) & (data.평일직장근처체류시간!=0) & (data.평일집근처체류시간!=0) & (data.카드소비금액!=0)
home=data.loc[condition__1] #이동횟수는 있지만 이동거리는 0 -> 기록안된것으로 간주
home  
```

```python
condition__2=((data.평일이동횟수==0) |(data.평일이동거리==0) | (data.휴일이동거리==0) | (data.평일직장근처체류시간==0) | (data.평일집근처체류시간==0)) & (data.카드소비금액!=0)
home_zeros=data.loc[condition__2]
home_zeros
```

```python
home_zeros['집순이집돌이지수']=3
```

```python
len(home), len(home_zeros)
```

```python
home_scaled=pd.DataFrame(data=home)
home_scaled['휴일평균이동거리']=home_scaled['휴일이동거리']/home_scaled['휴일이동횟수']
home_scaled['휴일가중치비율']=home_scaled['휴일직장근처체류시간']/home_scaled['평일직장근처체류시간']
home_scaled['집근처체류시간비율']=(home_scaled['평일집근처체류시간']+home_scaled['휴일집근처체류시간'])/(home_scaled['평일직장근처체류시간']+home_scaled['휴일직장근처체류시간'])
home_scaled
```

```python
home_zeros_scaled=pd.DataFrame(data=home_zeros)
home_zeros_scaled['휴일평균이동거리']=home_zeros_scaled['휴일이동거리']/home_zeros_scaled['휴일이동횟수']
home_zeros_scaled['휴일가중치비율']=home_zeros_scaled['휴일직장근처체류시간']/home_zeros_scaled['평일직장근처체류시간']
home_zeros_scaled['집근처체류시간비율']=(home_zeros_scaled['평일집근처체류시간']+home_zeros_scaled['휴일집근처체류시간'])/(home_zeros_scaled['평일직장근처체류시간']+home_zeros_scaled['휴일직장근처체류시간'])
home_zeros_scaled
```

```python
home_1=home_scaled.copy()
total_1=home_1['휴일이동횟수']
total_2=home_1['휴일평균이동거리']
total_3=home_1['집근처체류시간비율']
level_1q_1=total_1.quantile(0.25)
level_3q_1=total_1.quantile(0.75)
level_1q_2=total_2.quantile(0.25)
level_3q_2=total_2.quantile(0.75)
level_1q_3=total_3.quantile(0.25)
level_3q_3=total_3.quantile(0.75)
rev_range=3
IQR1=level_3q_1-level_1q_1
IQR2=level_3q_2-level_1q_2
IQR3=level_3q_3-level_1q_3
condition1=(total_1<=level_3q_1+(rev_range*IQR1))&(total_1>=level_3q_1-(rev_range*IQR1))
condition2=(total_2<=level_3q_2+(rev_range*IQR2))&(total_2>=level_3q_2-(rev_range*IQR2))
condition3=(total_3<=level_3q_3+(rev_range*IQR3))&(total_3>=level_3q_3-(rev_range*IQR3))
movement=home_1[condition1&condition2&condition3]
movement_else=home_1[(condition1==False)|(condition2==False)|(condition3==False)]
movement
```

```python
movement_fit=movement[['휴일이동횟수','휴일평균이동거리']].reset_index(drop=True)
movement_fit
movement_info=movement[home_scaled.columns.difference(['휴일이동횟수','휴일평균이동거리'])].reset_index(drop=True)
movement_info
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(movement_fit)
movement_scaled = scaler.transform(movement_fit)

movement_scaled_df = pd.DataFrame(data=movement_scaled, columns=['휴일이동횟수','휴일평균이동거리'])
movement_scaled_df
```

```python
homebody=pd.concat([movement_info,movement_scaled_df],axis=1)
homebody
```
